```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import mesa_reader as ms
```

```python
r = ms.MesaData('LOGS/history.data')

fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)

#Drop the pre-main-sequence calculation
ind = r.data('star_age') > 5e7
ax.plot(r.data('star_age')[ind]/1e9, r.data('log_L')[ind] + 3., lw=6,
        label="$\\log \\left( 10^3 \\times L_\star/L_\\odot\\right)$")
ax.plot(r.data('star_age')[ind]/1e9, r.data('log_Teff')[ind] - 3., ls='--', 
        label="$\\log \\left( T/{\\rm 10^3\ K}\\right)$")
ax.plot(r.data('star_age')[ind]/1e9, r.data('log_R')[ind] + 1., ls='-.',
        label="$\\log \\left( 10 \\times R_\star/R_\\odot\\right)$")
ax.legend(loc='best', fontsize=12)

ax.tick_params(axis='both', which='major', labelsize=24)
ax.set_xlabel('$t/{\\rm Gyrs}$', fontsize=24)

fig.savefig('Plot_MESA_Calculations_for_TRAPPIST-1.png', dpi=500, bbox_inches='tight')
```
